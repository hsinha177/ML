```
#importing library to download the web page from server
import requests as rq
#importing Beautiful Soup 4 libraray for beautifulsoup funtion which will help in parsing the web content
from bs4 import BeautifulSoup
```

```
#now sending the get request to server for GETting the webpage
page_download = rq.get('https://en.wikipedia.org/wiki/Tick')
```

```
#now viewing the content of the page downloaded
page_download.content
```

```
#now viewing the text of the webpage
page_download.text
```

```
#to see the status of the response fron server
page_download.status_code  #code starting with 2 shows success, 4 & 5 shows error in handeling the get request
```

```
#creating object of BeautifulSoup
soup=BeautifulSoup(page_download.content,'lxml')
```

```
#to see the text of the page downloaded
text_wp=soup.get_text()
```

```
# now print out the HTML content of the page, formatted nicely, using the prettify method on the BeautifulSoup object
soup.prettify()
```
